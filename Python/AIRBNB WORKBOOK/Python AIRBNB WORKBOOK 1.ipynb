### 21. Create a new column called hrt2  in which  string values in the host_response_time column are converted to integer values. 
 
* As you see below, the host_response_time column has only four values: "within an hour", "within a few hours", "with a day", "a few days or more", 
* Thus, you want to conver those four values to corresponding integers so that you can examine later whether fast response time is associated with higher review scores.

```python
replacement=df['host_response_time'].replace(to_replace={'within an hour':1,'within a few hours':2,'within a day':3,'a few days or more':4}).fillna(5)
df['hrt2']=replacement.astype('int64')
df[['host_response_time', 'hrt2']].head(10)
```

### 22. Create a new column called  host_veri_linkedin. If the host verification includes 'linkedin', the new column should show 1. Otherwise, 0. 

- You may create a custom funtion to return boolean(True/False) depending on whether the word "linkedin" is in the column. 
- After that, you have to use the Series apply method to call the function.   


```python
def containfunction(column1):
    if 'linkedin' in column1:
        return 1
    else:
        return 0
```

```python
df['host_veri_linkedin']=df['host_verifications'].apply(containfunction)
df[['host_verifications','host_veri_linkedin']].head(10)
```

### 24. How clean are Andrea's listings on average. Drop any missing value.


```python
import pandas as pd

#Make host name index for easy access
clean=pd.read_csv('Boston.csv', index_col='host_name')

rows = ['Andrea']
cols = 'review_scores_cleanliness'

#drop null values and assign to a variable
clean_drop = clean.loc[rows, cols].dropna()

#Calculate mean
clean_rating=clean_drop.mean()

print("Andrea's apartments have a rating of {} on average, which means her lisitngs are clean.".format(clean_rating))
```
